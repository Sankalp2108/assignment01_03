# 4. Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths',
#                                                 'gdp_per_capita',' human_development_index'] and update the data frame.

```python
df = pd.DataFrame(df,
                         columns = ['continent',
                                    'location',
                                    'date',
                                    'total_cases',
                                    'total_deaths',
                                    'gdp_per_capita',
                                    ' human_development_index']) 
print(df)
```

## 6. Date time format : 

## a. Convert date column in datetime format using pandas.to_datetime

```python
# a. Convert date column in datetime format using pandas.to_datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

## b. Create new column month after extracting month data from date column.

```python
# b. Create new column month after extracting month data from date column.

import datetime

df['month'] = pd.DatetimeIndex(df['date']).month
df.head()
```

# 7. Data Aggregation: 

## a. Find max value in all columns using groupby function on 'continent' column

```python
df2 = df.groupby(['continent']).max()
df2
```

## b. Store the result in a new dataframe named 'df_groupby'.

```python
# b. Store the result in a new dataframe named 'df_groupby'.

df_groupby = pd.DataFrame(df2)
df_groupby
```
